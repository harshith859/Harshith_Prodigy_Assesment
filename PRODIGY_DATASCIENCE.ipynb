```python
#Task 03
#Build a decision tree classifier to predict whether a customer will purchase a product
#or service based on their demographic and behavioral data. Use a dataset such as the Bank Marketing dataset from the UCI Machine Learning Repository.
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
file_path4 = '/content/sample_data/bank-full.csv'
DataFrame4 = pd.read_csv(file_path4, delimiter=';')
display(DataFrame4)

file_path5 = '/content/sample_data/bank.csv'
DataFrame5 = pd.read_csv(file_path5, delimiter=';')
display(DataFrame5)
```

```python
#Understand the data
#display the first 10 rows from data
display(DataFrame4.head())
display(DataFrame5.head())

#Check the data types and non-null counts
display(DataFrame4.info())
display(DataFrame5.info())

#Summarize the data
display(DataFrame4.describe(include='all'))
display(DataFrame5.describe(include='all'))
```

```python
#Data Preprocessing.
#Check for missing values
display(DataFrame4.isnull().sum())
display(DataFrame5.isnull().sum())
```

```python
#Since the data set might not have missing values , let's proceed with encoding.
#Convert catgorical variables to numeric using one-hot encoding
DataFrame4_encoded = pd.get_dummies(DataFrame4, drop_first=True)
DataFrame5_encoded = pd.get_dummies(DataFrame5, drop_first=True)

#For simplicity , let's assume y(Whether the customer subscribed) is our target variable.
x = DataFrame4_encoded.drop('y_yes', axis=1)
y = DataFrame4_encoded['y_yes']

a = DataFrame5_encoded.drop('y_yes', axis=1)
b = DataFrame5_encoded['y_yes']
```

```python
#Split the dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.2, random_state=42)
```

```python
#Build the decision tree classifier
#Initialize the classifier
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)

#Train the classifier
clf.fit(x_train, y_train)

#Make predictions
y_pred = clf.predict(x_test)

clf2 = DecisionTreeClassifier(random_state=42)
clf2.fit(a_train, b_train)
b_pred = clf2.predict(a_test)
```

```python
#Model evaluation
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100:.2f}%')

accuracy2 = accuracy_score(b_test, b_pred)
print(f'Accuracy: {accuracy2*100:.2f}%')

#Detailed Classification Report
print(classification_report(y_test, y_pred))

print(classification_report(b_test, b_pred))

#Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix2 = confusion_matrix(b_test, b_pred)
print('Confusion Matrix:\n',conf_matrix)
print('Confusion Matrix:\n',conf_matrix2)
```

```python
#Visualization
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

#Visualize the decision tree
plt.figure(figsize=(20, 10))
plot_tree(clf, filled=True, feature_names=x.columns, class_names=['No', 'Yes'], rounded=True)
plt.show()
```

```python
import pandas as pd
file_path6 = '/content/sample_data/twitter_training.csv'
DataFrame6 = pd.read_csv(file_path6)
display(DataFrame6)
```

```python
#understand the data
display(DataFrame6.head())
display(DataFrame6.info())
display(DataFrame6.describe())
```

```python
#Data Preprocessing
print(DataFrame6.isnull().sum)
DataFrame6.dropna(inplace=True)

#Convert text to lower case
DataFrame6['text'] = DataFrame6['text'].str.lower()
```

```python
import nltk
nltk.download('vader_lexicon')
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud

#Initialize the sentiment analyzer
sid = SentimentIntensityAnalyzer()

#Apply Sentimental Analysis to the text
DataFrame6['sentiment_score'] = DataFrame6['text'].apply(lambda x: sid.polarity_scores(x)['compound'])

#Classify Sentiment based on the score
DataFrame6['sentiment_label'] = DataFrame6['sentiment_score'].apply(lambda x: 'positive' if x > 0 else ('negative' if x < 0 else 'neutral'))

#Visualize the sentiment patterns
plt.figure(figsize=(10, 6))
sns.countplot(x='sentiment_label', data=DataFrame6)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
## 2. Sentiment Over Time (if timestamp is available)
# Ensure that the 'timestamp' column is properly formatted
if 'timestamp' in DataFrame6.columns:
    DataFrame6['timestamp'] = pd.to_datetime(DataFrame6['timestamp'])
    sentiment_over_time = DataFrame6.groupby([DataFrame6['timestamp'].dt.date, 'sentiment_label']).size().unstack()
    sentiment_over_time.plot(kind='line', stacked=False, figsize=(10, 6))
    plt.title('Sentiment Over Time')
    plt.xlabel('Date')
    plt.ylabel('Number of Tweets')
    plt.show()

# 3. Word Cloud for Each Sentiment
for sentiment in ['positive', 'negative', 'neutral']:
    sentiment_text = ' '.join(DataFrame6[DataFrame6['sentiment_label'] == sentiment]['text'].values)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(sentiment_text)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'{sentiment.capitalize()} Sentiment Word Cloud')
    plt.axis('off')
    plt.show()
```

```python
import pandas as pd
file_path7 = '/content/sample_data/US_Accidents.csv'
DataFrame7 = pd.read_csv(file_path7)
display(DataFrame7)
```

```python
#Explore the Dataset
display(DataFrame7.head())
display(DataFrame7.info())
display(DataFrame7.describe())
```

```python
#Data Cleaning and Preprocessing
print(DataFrame7.isnull().sum())

#For simplicity, lets drop the missing values
DataFrame7.dropna(inplace=True)
```

```python
# Convert relevant columns to datetime
DataFrame7['Start_Time'] = pd.to_datetime(DataFrame7['Start_Time'], format='%d-%m-%Y %H:%M')
DataFrame7['End_Time'] = pd.to_datetime(DataFrame7['End_Time'], format='%d-%m-%Y %H:%M')

# Extract hour, day, and month for further analysis
DataFrame7['Hour'] = DataFrame7['Start_Time'].dt.hour
DataFrame7['Day'] = DataFrame7['Start_Time'].dt.dayofweek
DataFrame7['Month'] = DataFrame7['Start_Time'].dt.month
```

```python
#Analyze the patterns
#1. Accidents by Time Of Day
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import folium
from folium.plugins import HeatMap
import numpy as np

# Visualize accidents by time of day
plt.figure(figsize=(10, 6))
sns.countplot(x='Hour', data=DataFrame7,palette='viridis')
plt.title('Accidents by Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
#Accidents by Month
plt.figure(figsize=(10, 6))
sns.countplot(x='Month', data=DataFrame7, palette='inferno')
plt.title('Accidents by Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
#Accidents by whether condition
# 4. Accidents by Weather Condition
plt.figure(figsize=(12, 8))
sns.countplot(y='Weather_Condition', data=DataFrame7, order=DataFrame7['Weather_Condition'].value_counts().iloc[:10].index, palette='magma')
plt.title('Top 10 Weather Conditions During Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Weather Condition')
plt.show()
```

```python
# Step 5: Visualize Accident Hotspots

# For the heatmap, we need to filter down the dataset to a manageable size
sampled_data = DataFrame7.sample(n=50, random_state=42)

# Create a map centered around the US
map = folium.Map(location=[37.0902, -95.7129], zoom_start=5)

# Add points to the map
heat_data = [[row['Start_Lat'], row['Start_Lng']] for index, row in sampled_data.iterrows()]
HeatMap(heat_data).add_to(map)

# Save the map as an HTML file
map.save('accident_heatmap.html')

```

```python
# Step 6: Analyze Contributing Factors

# 1. Accidents by Road Condition
plt.figure(figsize=(12, 8))
sns.countplot(y='Weather_Condition', data=DataFrame7, order=DataFrame7['Weather_Condition'].value_counts().iloc[:10].index, palette='cubehelix')
plt.title('Top 10 Road Conditions During Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Weather Condition')
plt.show()

# 2. Accidents by Visibility (in miles)
plt.figure(figsize=(10, 6))
sns.histplot(DataFrame7['Visibility(mi)'], bins=30, kde=True, color='purple')
plt.title('Distribution of Visibility During Accidents')
plt.xlabel('Visibility (miles)')
plt.ylabel('Frequency')
plt.show()

# 3. Accidents by Temperature (in Fahrenheit)
plt.figure(figsize=(10, 6))
sns.histplot(DataFrame7['Temperature(F)'], bins=30, kde=True, color='teal')
plt.title('Distribution of Temperature During Accidents')
plt.xlabel('Temperature (F)')
plt.ylabel('Frequency')
plt.show()

# 4. Accidents by Precipitation
plt.figure(figsize=(10, 6))
sns.histplot(DataFrame7['Precipitation(in)'], bins=30, kde=True, color='blue')
plt.title('Distribution of Precipitation During Accidents')
plt.xlabel('Precipitation (in)')
plt.ylabel('Frequency')
plt.show()
```
